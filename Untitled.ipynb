```python
from map.WorldMap import WorldMap
from map.land import City

import torch
import matplotlib.pyplot as plt
import numpy as np
from tqdm import trange
```

```python
np.random.seed(194)

device = "cuda"
map_generator_params = {
    "size": 2**10 + 1,  # Must be 2^n + 1 for diamond-square
    "scale": 1.0,
    "roughness": 0.55,
    "river_count": 5,
}

growth_factor = {
    "preference": [2.0, 1.2, 1.0],  # all land types prefer high fertility equally for now
    "proximity": [1,0, 1.0, 1.0],
    "distance": [4.0, 12.0, 1.0],
}

city_params = {
    "max_radius": 225,
    "max_road_radius" : 10,
    "growth_factor": growth_factor,
    
}

world_map = WorldMap(map_generator_params = map_generator_params, device = device, city_params=city_params, city_count=8)#, filename=True)

#world_map.saves()
#world_map.plot_all()
world_map.plot()
```

```python
world_map.cities[0].grow(amount = 100, land_type = 1, preference = "fertility")
```

```python
world_map.cities[0].grow(amount = 1000, preference = "fertility")
```

```python
from map.WorldMap import WorldMap

import torch
import matplotlib.pyplot as plt
import numpy as np
from tqdm import trange
```

```python

def random_between(land_type, preference):
    if land_type == 0:
        if preference == "fertility":
            a, b = 500, 1500
        else:
            a, b = 250, 1000
    else:
        a, b = 30, 150
    return int(a + (b - a) * np.random.rand())

def grow(index, land_type, preference, area):
    rnd = random_between(land_type, preference)
    world_map.cities[index].grow(amount = rnd, land_type = land_type, preference = preference)
    area[index, land_type] += rnd
```

```python
np.random.seed(200)

device = "cuda"
map_generator_params = {
    "size": 2**10 + 1,  # Must be 2^n + 1 for diamond-square
    "scale": 1.0,
    "roughness": 0.55,
    "river_count": 5,
}

growth_factor = {
    "preference": [1.5, 1.0, 1.0],  # all land types prefer high fertility equally for now
    "distance": [4, 8, 1.0],
}

city_params = {
    "max_radius": 225,
    "max_road_radius" : 20,
    "growth_factor": growth_factor,
    
}

world_map = WorldMap(map_generator_params = map_generator_params, device = device, city_params=city_params, city_count=8, filename=True)
cities = 2
area = np.zeros((cities,2))
#world_map.save()
#world_map.plot_all()
#world_map.plot()
```

```python

for i in trange(100):
    for index in range(cities):
        for preference in ["fertility", "forest"]:
            for land_type in range(2):
                grow(index, land_type=land_type, preference=preference, area=area)
                

```

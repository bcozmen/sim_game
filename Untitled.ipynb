```python
from map.WorldMap import WorldMap
from map.land import City
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python

np.random.seed(70)



device = "cuda"
map_generator_params = {
    "size": 2**13 + 1,  # Must be 2^n + 1 for diamond-square
    "scale": 1.0,
    "roughness": 0.45,
    "river_count": 10,
}

city_params = {
    "max_radius": 2000,
    "growth_factor": {
        0: (1.0, -1.0),  # urban areas prefer high fertility and proximity
        1: (1.0, -0.5),  # farmland prefers high fertility but less proximity penalty
        2: (1.0, -0.2)   # industrial areas prefer high fertility but even less proximity penalty
    }
}

world_map = WorldMap(map_generator_params = map_generator_params, device = device, city_params=city_params, city_count=8)#, filename=True)

#world_map.save_maps()
world_map.plot_maps('height')
```

```python
world_map.maps['roads'].shape
```

```python
growth_factor = {
    0: (1.5, 3),  # a, b for growth score calculation
    1: (1.5, 12),
}
city = City(id=1, maps=world_map.maps, pos=(2500,2500), max_radius=500, growth_factor= growth_factor)
```

```python
preference = "fertility"
world_map.plot_maps(preference)
city.grow(preference = preference)
```
